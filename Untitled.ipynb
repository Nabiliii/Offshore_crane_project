# Extract the bounding box information and save them to new file:

```python
import json

def extract_bbox_info(json_path, output_path):
    with open(json_path, 'r') as json_file:
        data = json.load(json_file)

    bounding_boxes = []

    for idx, capture in enumerate(data['captures']):
        for annotation in capture['annotations']:
            if 'values' in annotation and len(annotation['values']) > 0:
                bbox_info = annotation['values'][0]
                bounding_boxes.append({
                    'filename': f'img{idx}',
                    'bbox': (bbox_info['x'], bbox_info['y'], bbox_info['width'], bbox_info['height'])
                })

    with open(output_path, 'w') as output_file:
        json.dump(bounding_boxes, output_file, indent=2)

# Example usage with raw strings
json_path = r"C:\Users\abdak\OneDrive\Skrivebord\perception\dataset\Datasetdaa83fdc-4299-402a-b533-a1bc284144c7\captures_000.json"
output_path = r"C:\Users\abdak\OneDrive\Skrivebord\perception\dataset\annotations.json.txt"

extract_bbox_info(json_path, output_path)
```

# plot an example img and corres. bounding box


```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_image_with_bbox(image_path, bbox):
    # Read the image
    image = cv2.imread(image_path)

    # Convert BGR to RGB (OpenCV uses BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(image_rgb)

    # Extract the coordinates of the bounding box
    x, y, w, h = bbox

    # Create a rectangle patch
    rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='r', facecolor='none')

    # Add the rectangle to the axes
    ax.add_patch(rect)

    # Show the plot
    plt.show()

# Example usage
image_pathh = "C:\\Users\\abdak\\OneDrive\\Skrivebord\\perception\\dataset\\images\\img156.png"

bounding_boxx =(
      944.0,
      656.0,
      91.0,
      410.0
)
plot_image_with_bbox(image_pathh, bounding_boxx)
```

```python
import os

def rename_files(folder_path):
    files = sorted(os.listdir(folder_path))

    for idx, filename in enumerate(files):
        new_name = f'img{idx}{os.path.splitext(filename)[1]}'
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        os.rename(old_path, new_path)

# Example usage
folder_path = r"C:\\Users\\abdak\\OneDrive\\Skrivebord\\perception\\dataset\\images"
rename_files(folder_path)
```
