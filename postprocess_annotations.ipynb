# Extract the bounding box information and save them to new file:

```python
import json

def extract_bbox_info(json_path, output_path):
    with open(json_path, 'r') as json_file:
        data = json.load(json_file)

    bounding_boxes = []

    for idx, capture in enumerate(data['captures']):
        for annotation in capture['annotations']:
            if 'values' in annotation and len(annotation['values']) > 0:
                bbox_info = annotation['values'][0]
                bounding_boxes.append({
                    'filename': f'img{idx}',
                    'bbox': (bbox_info['x'], bbox_info['y'], bbox_info['width'], bbox_info['height'])
                })

    with open(output_path, 'w') as output_file:
        json.dump(bounding_boxes, output_file, indent=2)

# Example usage with raw strings
json_path = r"C:\Users\abdak\OneDrive\Skrivebord\perception\dataset1\annot\captures_000.json"
output_path = r"C:\Users\abdak\OneDrive\Skrivebord\perception\dataset1\annotations.json"

extract_bbox_info(json_path, output_path)
```

# plot an example img and corres. bounding box


```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_image_with_bbox(image_path, bbox):
    # Read the image
    image = cv2.imread(image_path)

    # Convert BGR to RGB (OpenCV uses BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(image_rgb)

    # Extract the coordinates of the bounding box
    x, y, w, h = bbox

    # Create a rectangle patch
    rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='r', facecolor='none')

    # Add the rectangle to the axes
    ax.add_patch(rect)
    ax.axis ('off')
    # Show the plot
    plt.show()

# Example usage
image_pathh = "C:\\Users\\abdak\\OneDrive\\Skrivebord\\perception\\dataset1\\images\\rgb_4233.png"


bounding_boxx =(
      673.0,
      155.0,
      540.0,
      925.0
)
plot_image_with_bbox(image_pathh, bounding_boxx)


#ecample 2
# Example usage
image_path1 = "C:\\Users\\abdak\\OneDrive\\Skrivebord\\perception\\dataset1\\images\\rgb_1167.png"


bounding_box1 =(
      936.0,
      507.0,
      272.0,
      250.0
)
plot_image_with_bbox(image_path1, bounding_box1)



#ecample 3
# Example usage
image_path2 = "C:\\Users\\abdak\\OneDrive\\Skrivebord\\perception\\dataset1\\images\\rgb_1524.png"


bounding_box2 =(
      1164.0,
      343.0,
      437.0,
      178.0
)
plot_image_with_bbox(image_path2, bounding_box2)
```
